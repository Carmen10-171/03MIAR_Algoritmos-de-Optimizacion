## 01- 1.25pt
- Dado un diccionario con una lista de nombres de películas y al año concatenado, crear una función que devuelva un DataFrame con dos columnas:
  - Columna 1, el nombre de la película sin el año
  - Columna 2, el año de la película

```python
# Your solution HERE !!!!
def extract_year(source_movies):

    names = []
    years = []

    for movie in source_movies['Name']:

        movie_split = movie.split(' - ')
        names.append(movie_split[0])
        years.append(movie_split[1])


    data_frame = pd.DataFrame({
        'Name' : names,
        'Year' : years,
    })

    return data_frame
```

```python
# Test Use Case
source_movies = {'Name': ["Matrix - 1999", "Cadena Perpetua - 1994", "Parasitos - 2019", "Vengadores, Endgame - 2019", "Interstellar - 2014",
                   "Django Desencadenado - 2012", "El Padrino - 1972", "El Caballero Oscuro - 2008", "Erase Una Vez En America - 1984",
                   "El Senyor de los Anillos, La Comunidad del Anillo - 2001"]}

expected_output = pd.DataFrame({'Name':["Matrix", "Cadena Perpetua", "Parasitos", "Vengadores, Endgame", "Interstellar", "Django Desencadenado",
                                        "El Padrino", "El Caballero Oscuro", "Erase Una Vez En America", "El Senyor de los Anillos, La Comunidad del Anillo"],
                                'Year':["1999", "1994", "2019", "2019", "2014", "2012", "1972", "2008", "1984", "2001"]})

output = extract_year(source_movies)   # This will call your function extract_year

display(expected_output)
display(output)

assert(expected_output.equals(output))  # This will fail if the result is not as expected
```

## 03 - 1.25 pt
- Dado un array de elementos, escribir una función, que reciba dicho array como parámetro de entrada y devuelva el elemento repetido. Usando plain Python, sin ninguna librería.

```python
# Your solution HERE !!!!
def repeated_element(array):
    checked_numbers = []
    repeated_number = 0

    for num in array:
        if num in checked_numbers:
            repeated_number = num
            break

        checked_numbers.append(num)

    return repeated_number
```

```python
# Test Use Case
source_array = [4, 8, 0, 3, 6, 2, 1, 5, 3, 9, 7]
expected_rep_element = 3

rep_element = repeated_element(source_array) # This calls your function repeated_element

print(rep_element)

assert(expected_rep_element == rep_element)  # This will fail if the element is not found correctly
```

## 04 - 1.25pt
- Resolución de dos sistemas de ecuaciones
    - $ A : \begin{cases} 8x +  2y = 11 \\ -3y + 10x = 44                         \end{cases}$
    
    - $ B : \begin{cases} -4z + y = 5 \\ 3y + 2z = -2 \\ -x -z + 2y = -1 \end{cases}$
- Verifica por código los resultados

```python
# Your solution HERE !!!!
A = np.array([
    [8, 2],
    [10, -3]
])

result_A = np.array([11, 44])

resultado_A = np.linalg.solve(A, result_A)

print("Resultado sistema A:")
print("x =", resultado_A[0])
print("y =", resultado_A[1])


# Verificación sistema A
print("Verificación A:")
print(8 * resultado_A[0] + 2 * resultado_A[1])
print(10 * resultado_A[0] - 3 * resultado_A[1])



B = np.array(
    [
     [0,  1, -4],
     [0,  3,  2],
     [-1, 2, -1],
    ])

result_B = np.array([5, -2, -1])

resultado_B = np.linalg.solve(B, result_B)

x = resultado_B[0]
y = resultado_B[1]
z = resultado_B[2]

print("Resultado sistema B:")
print("x =", x)
print("y =", y)
print("z =", z)

print("Verificación B:")
print(-4*z + y)
print(3*y + 2*z)
print(-1*x  -1*z + 2*y)
```

```python
# 1.A. Muestra las seis primeras filas.
df_aircraft = pd.read_csv('aircraft.csv')

display(df_aircraft.head(6))
```

```python
# 1.B. Calcula el porcentaje de valores nulos por columna.

porcentaje_null_sum = df_aircraft.isnull().sum()
porcentaje_null = ( porcentaje_null_sum / len(df_aircraft) ) * 100
display(porcentaje_null)
```

```python
# 1.C. Descarta del df original los registros cuyos valores de latitud y longitud no se encuentren en el rango
# de coordenadas esféricas, siendo la longitud valida en el rango [-180, 180], y la latitud valida en el rango [-90, 90]

df_aircraft = df_aircraft[
                (df_aircraft['LONGITUDE'] >= -180) &
                (df_aircraft['LONGITUDE'] <= 180) &
                (df_aircraft['LATITUDE'] >= -90) &
                (df_aircraft['LATITUDE'] <= 90)
            ]
display(df_aircraft)
```

```python
# 1.D. Muestra en una figura, con un tamaño de 8 de ancho y 4 de alto, una gráfica de tipo puntos (scatter),
# la localización de los incidentes en base a la longitud y latitud (ya limpios los rangos invalidos en el punto anterior).

plt.figure(figsize=(8,4))
x = df_aircraft['LONGITUDE']
y = df_aircraft['LATITUDE']

plt.scatter(x,y, s=3)
```

```python
# 1.E. Muestra en una figura un conteo de incidentes sobre la columna 'AIRCRAFT' por año, en el gráfico debe aparecer
# una columna por cada 'DAMAGE_LEVEL' distinto en el dataframe. La gráfica debe ser de tipo linea con diferentes subplots.
# Pista, agrupa primero y aplica una gráfica de relación.


df_plot = df_aircraft.dropna(subset=['INCIDENT_DATE', 'DAMAGE_LEVEL', 'AIRCRAFT'])

grouped = df_plot.groupby(
    ['INCIDENT_DATE', 'DAMAGE_LEVEL']
)['AIRCRAFT'].count().unstack(fill_value=0)

grouped.plot(
    kind='line',
    subplots=True,
    figsize=(10, 8),
    title='Conteo de incidentes por año según DAMAGE_LEVEL'
)

plt.show()

```

```python
# 2.A. Muestra el número de filas y columnas

df_airports = pd.read_csv('airports.csv')
display(df_airports.shape)
display(df_airports.head(6))
```

```python
# 2.B. Cambia los valores a minúsculas de columna de nombre de aeropuerto en este dataframe y en el del apartado anterior.
df_airports['name'] = df_airports['name'].str.lower()
df_aircraft['AIRPORT'] = df_aircraft['AIRPORT'].str.lower()

display(df_airports.head(6))
display(df_aircraft.head(6))
```

```python
# 2.C. Con el propósito de disponer del país de los aeropuertos, junta ambos dataframes para obtener un nuevo dataframe
# con todos los registros de incidentes de avión y los países de los aeropuertos. Como clave, utilizar el nombre del
# aeropuerto modificado en el anterior apartado.

print(df_aircraft['AIRPORT'].duplicated().sum())
print(df_airports['name'].duplicated().sum())

df_airports_unique = df_airports.drop_duplicates(subset=['name'])

df_join = pd.merge(
            df_aircraft,
            df_airports_unique,
            left_on='AIRPORT',
            right_on='name',
            how='left'
        )
```

```python
# 2.D. Como resultado se debe disponer un dataframe con 254222 registros.
print(len(df_join))
```

```python
# 2.E. Muestra el número de registros que no disponen de dato en el atributo "iso_country".
registros_sin_iso_country = df_join['iso_country'].isna().sum()
print(f'Numero de registros sin iso_country = {registros_sin_iso_country} ')
```

```python
# 2.F. Elimina todos los registros que que tengan nulos alguna de las columnas 'DAMAGE_LEVEL' y 'iso_country'.
df_clean_join = df_join.dropna(subset=['DAMAGE_LEVEL', 'iso_country'])

registros_sin_iso_country = df_clean_join['iso_country'].isna().sum()
print(f'Numero de registros sin iso_country = {registros_sin_iso_country} ')

registros_sin_damage_level= df_clean_join['DAMAGE_LEVEL'].isna().sum()
print(f'Numero de registros sin DAMAGE_LEVEL = {registros_sin_damage_level} ')
```

```python
# 2.G. Sobre el resultado anterior, crea una tabla contando el número de incidentes por país, en las filas deben
# aparecer todos los paises posibles y en las columnas debe de estar el tipo de 'DAMAGE_LEVEL'.

tabla = pd.crosstab(
    df_clean_join['iso_country'],
    df_clean_join['DAMAGE_LEVEL']
)

display(tabla)

```
